```python
import requests 
url = "http://127.0.0.1:8001/api/login/"
```

```python
login_data = {
    'username' : 'user1',
    'password' : 'password123'
}
```

```python
r = requests.post(url, json=login_data)
```

```python
r.status_code
```

```python
token = r.json()['token']
```

```python
headers = {
    'Authorization' : f"Token {token}",
    "Content-type" : "application/json"
}
```

```python
headers
```
