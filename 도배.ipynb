```python
import requests
```

```python
url_login = "http://127.0.0.1:8000/api/login/"
login_data = {
    'username' : 'seo',
    'password' : '123'
}
login_response = requests.post(url_login, json=login_data)
print (login_response.status_code)
```

```python
head = {
    "Authorization" : f"Bearer {login_response.json()['token']}",
    "Content-Type": "application/json",
}
```

```python
question_list_url = "http://127.0.0.1:8000/api/questions/"
response = requests.get(question_list_url , headers=head)
```
